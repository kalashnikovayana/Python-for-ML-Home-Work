2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp

x = sp.symbols('x')

# Функції
f1 = sp.sin(x)
f2 = sp.ln(x)

# Похідні
df1 = sp.diff(f1, x)
df2 = sp.diff(f2, x)

print("f(x) = sin(x) → похідна:", df1)
print("f(x) = ln(x) → похідна:", df2)
```

```python
f1_func = sp.lambdify(x, f1, modules=['numpy'])   # sin(x)
df1_func = sp.lambdify(x, df1, modules=['numpy']) # cos(x)

f2_func = sp.lambdify(x, f2, modules=['numpy'])   # ln(x)
df2_func = sp.lambdify(x, df2, modules=['numpy']) # 1/x
```

```python
xs = np.linspace(0.1, 10, 200)

plt.figure(figsize=(10, 6))

# sin(x) та її похідна
plt.plot(xs, f1_func(xs), label='f(x) = sin(x)', color='blue')
plt.plot(xs, df1_func(xs), '--', label="f'(x) = cos(x)", color='skyblue')

# ln(x) та її похідна
plt.plot(xs, f2_func(xs), label='f(x) = ln(x)', color='green')
plt.plot(xs, df2_func(xs), '--', label="f'(x) = 1/x", color='lime')

plt.title("Функції та їхні похідні")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x = sp.symbols('x')
function = x**2 - 5 * x + 1
derivative = sp.diff(function, x)

# Перетворюємо у звичайні Python-функції
f = sp.lambdify(x, function, 'numpy')
df = sp.lambdify(x, derivative, 'numpy')

# Відрізок від 0 до 100
xs = np.linspace(0, 100, 300)

# Побудова графіка
plt.figure(figsize=(10, 6))
plt.plot(xs, f(xs), label='f(x) = x² - 5x + 1', color='blue')
plt.plot(xs, df(xs), '--', label="f'(x) = 2x - 5", color='orange')

plt.title("Функція витрат і її похідна")
plt.xlabel("Кількість одиниць продукції (x)")
plt.ylabel("Витрати")
plt.legend()
plt.grid(True)
plt.show()
```

Спостереження:

При збільшенні обсягу виробництва витрати зростають нелінійно.
Це означає, що кожна додаткова одиниця продукції коштує дорожче, ніж попередня.
